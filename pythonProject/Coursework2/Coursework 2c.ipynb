```python
#QUESTION C:  Clean and transform the data into the correct data types
import pandas as pd
file_path = 'C:\\Users\\atang\\OneDrive\\Documents\\Data Science MSc\\Programming for data science\\who_suicide_statistics_modifed3 (2).csv'
Dataset = pd.read_csv(file_path)
#remove whitespaces in column names to make it easier when writing them in subsequent codes and avoid errors in code outputs
Dataset.columns = Dataset.columns.str.strip()
#check if there are still whitespaces to make sure the code above was applied
print(Dataset.columns)

#remove whitespaces from string columns only, to allow better subsequent cleaning results
for column in Dataset.columns:
    if Dataset[column].dtype == 'object': #object meaning text, non-numerical and numerical values
        Dataset[column] = Dataset[column].str.strip() 

#Convert object columns based on conclusion in question b of data types to numerical values
    #convereting to numeric values results in missing values to be converted to NaN
    #from here, it will be easier to fill in missing values (NaN) using .fillna
#Write a function to identify columns with data types = object and perform the conversion to numerical values
def safe_convert_to_numeric(column):
    if column.dtype == object:
        return pd.to_numeric(column.str.replace('[^0-9.]', '', regex=True), errors='coerce')
    else:
        return column


Dataset['suicides_no'] = safe_convert_to_numeric(Dataset['suicides_no'])
Dataset['gdp_for_year ($)'] = safe_convert_to_numeric(Dataset['gdp_for_year ($)'])

#Handling missing values
#Fill in missing values with median as it is not affected by outliers
Dataset['suicides_no'].fillna(Dataset['suicides_no'].median(), inplace = True) 
Dataset['HDI for year'].fillna(Dataset['HDI for year'].median(), inplace = True) 

#check data types have been converted
print(Dataset['suicides_no'].dtype)
print(Dataset['population'].dtype)
print(Dataset['HDI for year'].dtype)
print(Dataset['gdp_for_year ($)'].dtype)
print(Dataset['population'])

#check if they are any more missing values to make sure the missing rows haven been filled in with median values
Datasetnew = Dataset.isnull().sum()
print(Datasetnew)

#Handling outliers
#checking for outliers and removing them improves the accuracy of the results of our data analysis
#Calculate the Upper, Lower Quartile and Interquartile range. Use the formular Q3 + or - 1.5*IQR  to identify outliers
Q1 = Dataset['suicides_no'].quantile(0.25)
Q3 = Dataset['suicides_no'].quantile(0.75)
IQR = Q3 -Q1

lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
outliers_options = (Dataset['suicides_no'] < lower_bound) | (Dataset['suicides_no'] > upper_bound)

outliers = Dataset['suicides_no'][outliers_options]
print(outliers) #4144 outliers identified. they were not treated as noisy data as they could indicate countries with high suicide rates.

#the following code below would have been used to drop/remove outliers from the dataset if deemed so
#Dataset['suicides_no'] = Dataset['suicides_no'][~((Dataset['suicides_no'] < (Q1 - 1.5*IQR)) | (Dataset['suicides_no'] > (Q3 + 1.5*IQR)))]

#Handling duplicates
#Drop duplicates and keep the last occurrence
#drop duplicates 
Dataset.drop_duplicates(keep = 'last', inplace = True)

#check if there are any more duplicates that may have been missed 
print(Dataset.duplicated().sum())

#Display the first few lines of the dataset with filled values to see if the code above has been processed
Dataset_head = Dataset.head()
print(Dataset_head)

#Display length of dataset#Display length of dataset
print(Dataset.shape)

#save cleaned dataset to a new cvs file to make sure we are analysing cleaned data. This will avoid inaccuracy and inconsistencies within the data
Dataset_cleaned_filepath = 'C:\\Users\\atang\\OneDrive\\Documents\\Data Science MSc\\Programming for data science\\cleaned_who_suicide_statistics_modified.csv'
Dataset.to_csv(Dataset_cleaned_filepath, index = False) #If index=False, the CSV file will not have a column for the index, and it starts directly with the data columns.
```
